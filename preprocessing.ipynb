## **데이터 저장**

````python
import datas

chats_txt = datas.dasom_japdam_txt
to_csv_name = datas.dasom_japdam_csv

chats = open(chats_txt, 'r', encoding='utf8')
chats_content = chats.readlines()
chats.close()
````

````python
print(chats_content)
````

## **'\n' 전처리**

````python
for i, _ in enumerate(chats_content):
    chats_content[i] = chats_content[i].strip('\ufeff').strip('\n')

print(chats_content)
````

```python
['2025년 3월 3일 오전 10:50, 홍성학 : @조성언 성언이랑 저랑 점심 같이먹을 신입생분들 구해요~', '맛있는거 사드릴게요 ㅠ']
```

```bash
____년 _월 _일 __ __:__
```
만큼을 날짜 정규표현식으로,

```bash
, {이름} : {내용}
```
만큼을 내용 정규표현식으로 나타냈습니다.

````python
import re

date_pattern = r"^\d{4}년 \d{1,2}월 \d{1,2}일 [가-힣]+ \d{1,2}:\d{1,2}"
content_pattern = r", [가-힣a-zA-Z]+ : .*"
````

날짜 정규표현식으로 시작하지 않는 인덱스는 앞쪽 인덱스에 붙여줬습니다.

````python
i=1
while i<len(chats_content):
    if (not re.search(date_pattern, chats_content[i])):
        chats_content[i - 1] += ' ' + chats_content[i]
        chats_content.pop(i)
        i -= 1

    i += 1

print(chats_content)
````

날짜 정규표현식 + 내용 정규표현식 : 
```go
~~년 ~~월 ~~일 오후/오전 ~~:~~, {이름} : {내용}
```
의 형태가 아닌 인덱스를 전부 삭제해줬습니다.

````python
i=0
while i<len(chats_content):
    if (not re.search(date_pattern+content_pattern, chats_content[i])):
        chats_content.pop(i)
        i -= 1

    i += 1

print(chats_content)
````

## **CSV로 저장**

````python
date, name, context = [], [], []

for idx, value in enumerate(chats_content):
    
    date.append(value[:value.index(', ')])
    name.append(value[value.index(', ')+2:value.index(':', value.index(':') + 1)-1])
    context.append(value[value.index(':', value.index(':') + 1)+2:])


map(str, context)

for i in range(len(name)):
    print(date[i], name[i], context[i], sep='***')
````

````python
from datetime import datetime

for idx, value in enumerate(date):
    value = value.replace('오전', "AM")
    value = value.replace('오후', "PM")

    date[idx] = datetime.strptime(value, "%Y년 %m월 %d일 %p %I:%M")
````

````python
print(len(date), len(name), len(context))
````

````python
import pandas as pd

df = pd.DataFrame({"date":date, "name":name, "context":context})

df.to_csv(to_csv_name, index=False)
````
